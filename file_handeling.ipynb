```python
# context manager: helps in automation of file handling
# this is generay way of file handling

read_file = open(r'/Users/varun/Desktop/python 2026\fake_file.txt','r')
print(read_file.read())
read_file.close()
```

```python
read_file = open(r'/Users/varun/Desktop/python 2026\fake_file.txt','w')
# this is using try, except, finally. it will try to write it, if there is an error it will print e, or else it will close it
try:
    read_file.write('Hey!!')
except exception as e:
    print(e)
finally:
    read_file.close()
```

```python
with open(r'/Users/varun/Desktop/python 2026\fake_file.txt','w') as read_file:
    read_file.write('I am using context manager ')  # using context manager
```

```python
import csv

col_name = ['name', 'job', 'age']

data = [['var', 'student', 25], ['par', 'clerk', 23], ['kar', 'front desk', 26]]

with open(r'/Users/varun/Desktop/python 2026\csv_file.csv','w') as write_file:
    csv_writer = csv.writer(write_file)

    csv_writer.writerow(col_name)

    csv_writer.writerows(data)
```
